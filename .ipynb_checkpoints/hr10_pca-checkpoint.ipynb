# In what follows:
I am worried I made a mistake from the very linear relationship between the first and second PCA coefficients for each spectra.

I believe the reconstruction is working from the above analysis. The functional form of this is : 
$$ \vec{x}_i=\vec{\mu}+(AB)_i$$
where the index $i$ corresponds to a row in the matrix learning_spectra. Moreover, $A$ is the 1080 by 10 matrix spec_fitted, $B$ is the 10 by 5541 matrix pca.pca.components_.    

The above equation seems to be able to succesfuly reconstruct the spectra. Below I plot the first column of A against the second column of A.

```python
pca_coeff=pd.DataFrame(data=pca.spec_fitted)
pca_analysis=pd.concat([analysis,pca_coeff],axis=1)
pca_analysis=pca_analysis[pca_analysis['LOGG']!='nan'].astype(float)

plt.figure(figsize=(15,10))
plt.scatter(x=pca_analysis[0],y=pca_analysis[1],s=0.3)
plt.xlim(-80000,50000)
plt.ylim(-15000,20000)
plt.title('Plot of spectra components')
plt.ylabel('Second PCA component')
plt.xlabel('First PCA Component')
```

```python
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')

sequence_containing_x_vals = pca.spec_fitted.transpose()[0]
sequence_containing_y_vals = pca.spec_fitted.transpose()[1]
sequence_containing_z_vals = pca.spec_fitted.transpose()[2]

ax.set_xlim(-80000,50000)
ax.set_ylim(-15000,20000)

ax.scatter(sequence_containing_x_vals, sequence_containing_y_vals, sequence_containing_z_vals)
plt.show()
```
